##折れ線グラフ

```python
import numpy as np

x = np.arange(0.0, 15.0, 0.1) #0.0~15.0、0.1間隔
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
ax.legend()

plt.show()
```

##棒グラフ

```python
fix, ax = plt.subplots()

x = [1, 2, 3]
y = [10, 2, 3]
ax.bar(x, y)

plt.show()
```

```python
fix, ax = plt.subplots()

x = [1, 2, 3]
y = [10, 2, 3]
labels = ['spam', 'ham', 'egg']
ax.bar(x, y, tick_label=labels) #ラベルを指定

plt.show()
```

```python
#横向きの棒グラフを描画
fix, ax = plt.subplots()

x = [1, 2, 3]
y = [10, 2, 3]
labels = ['spam', 'ham', 'egg']
ax.barh(x, y, tick_label=labels) #ラベルを指定

plt.show()
```

```python
fix, ax = plt.subplots()

x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spam', 'ham', 'egg']

width = 0.4
ax.bar(x, y1, width=width, tick_label=labels, label='y1') #ラベル、幅を指定して描画

#幅をずらして棒グラフを描画
x2 = [num + width for num in x]
ax.bar(x2, y2, width=width, label='y2')

ax.legend()

plt.show()
```

##積み上げ棒グラフ

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spam', 'ham', 'egg']

# y1とy2を足した値を格納
y_total = [num1 + num2 for num1, num2 in zip(y1, y2)]

# y1とy2を足した高さの棒グラフを描画
ax.bar(x, y_total, tick_label=labels, label='y1')
# かたほうの値のみ棒グラフを描画
ax.bar(x, y2, label='y2')
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

# ランダムに50個の要素を生成
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x[0:10], y[0:10], marker='v', label='triangle down') #下向き三角
ax.scatter(x[10:20], y[10:20], marker='^', label='triangle up') #上向き三角
ax.scatter(x[20:30], y[20:30], marker='s', label='square') #正方形
ax.scatter(x[30:40], y[30:40], marker='*', label='star') #星型
ax.scatter(x[40:50], y[40:50], marker='x', label='x') #X
ax.legend()

plt.show()
```

##ヒストグラム

```python
import numpy as np
import matplotlib.pyplot as plt

# データ生成
np.random.seed(123)
mu = 100 #平均値
sigma = 15 #標準偏差
x = np.random.normal(mu, sigma, 1000) # 정규분포로부터 난수 생성 (평균: mu, 표준편차: sigma, 샘플 수: 1000)


fig, ax = plt.subplots()

#ヒストグラムを描画
n, bins, patches = ax.hist(x) 
# n: 各ビン（棒）の度数（要素数）
# bins: ビンの境界の値
# patches: ビンを描画するための情報

plt.show()
```

```python
for i, num in enumerate(n):
    print('{:,.2f} - {:,.2f}: {}'.format(bins[i],  bins[i + 1], num))
```

```python
fig, ax = plt.subplots()

ax.hist(x, bins=25) #ビンの数を指定して表示

plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(x, orientation='horizontal') #ビンの数を指定して表示

plt.show()
```

```python
#データ生成
np.random.seed(123)
mu = 100 #平均値
x0 = np.random.normal(mu, 20, 1000)

#異なる標準偏差でデータを作成
x1 = np.random.normal(mu, 15, 1000)
x2 = np.random.normal(mu, 10, 1000)

fig, ax = plt.subplots()

labels = ['x0', 'x1', 'x2']
#3つのデータのヒストグラムを描画
ax.hist((x0, x1, x2), label=labels)
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

labels = ['x0', 'x1', 'x2']
#3つのデータのヒストグラムを描画
ax.hist((x0, x1, x2), label=labels, stacked=True) #stack=Trueと指定すると積み上げヒストグラムが描画される
ax.legend()

plt.show()
```

##箱ひげ図

```python
import numpy as np
import matplotlib.pyplot as plt

#　異なる標準偏差でデータを生成
np.random.seed(123)
x0 = np.random.normal(0, 10, 500)
x1 = np.random.normal(0, 15, 500)
x2 = np.random.normal(0, 20, 500)

fig, ax = plt.subplots()
labels = ['x0', 'x1', 'x2']
ax.boxplot((x0, x1, x2), labels=labels)

plt.show()
```

```python
fig, ax = plt.subplots()
labels = ['x0', 'x1', 'x2']
# 横向きの箱ひげ図を描画
ax.boxplot((x0, x1, x2), labels=labels, vert=False)

plt.show()
```

##円グラフ

```python
labels = ['spam', 'ham', 'egg']
x = [10, 3, 1]

fig, ax = plt.subplots()

ax.pie(x, labels=labels)
# アスペクト比を保持して描画する
ax.axis('equal')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.pie(x, labels=labels, startangle=90, counterclock=False) #上から時計回り
ax.axis('equal')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.pie(x, labels=labels, startangle=90, counterclock=False, #上から時計回り
      shadow=True, autopct='%1.2f%%') #影、％表示指定
ax.axis('equal')

plt.show()
```

```python
explode = [0, 0.2, 0] #1番目のhamを取り出す

fig, ax = plt.subplots()

ax.pie(x, labels=labels, startangle=90, counterclock=False, #上から時計回り
      shadow=True, autopct='%1.2f%%', #影、％表示指定
      explode=explode) #explode指定
ax.axis('equal')

plt.show()
```

##複数のグラフの組み合わせ方

```python
fig, ax = plt.subplots()

x1 = [1, 2, 3]
y1 = [5, 2, 3]
x2 = [1, 2, 3, 4]
y2 = [8, 5, 4, 6]
ax.bar(x1, y1, label='y1') #棒グラフ描画
ax.plot(x2, y2, label='y2') #折れ線グラフを描画
ax.legend()

plt.show()
```

```python
np.random.seed(123)
x = np.random.randn(1000)

fig, ax = plt.subplots()

# ヒストグラムを描画
counts, edges, patches = ax.hist(x, bins=25)

# 近似曲線い用いる点を求める（ヒストグラムのビンの中点）
x_fit = (edges[:-1] + edges[1:]) / 2
# 近似曲線をプロット
y = 1000 * np.diff(edges) * np.exp(-x_fit**2 / 2) / np.sqrt(2 * np.pi)
ax.plot(x_fit, y)

plt.show()
```

##スタイル

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# 線の色を名前で指定
ax.plot([1, 3], [3, 1], label='aqua', color='aqua') #HTML, X11, CSS4で定義された色名
# 16進数のRGBで指定
ax.plot([1, 3], [1, 3], label='#0000FF', color='#0000FF')
# RGBAをfloatで指定
ax.plot([1, 3], [2, 2], label='(0.1, 0,2, 0.5, 0.3)', color=(0.1, 0.2, 0.5, 0.3))
ax.legend() #凡例を表示

plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar([1], [3], color='aqua') # 塗りつぶしの色を指定
#塗りつぶしの色と枠線の色を指定
ax.bar([2], [4], color='aqua', edgecolor='black')

plt.show()
```

##線のスタイル

```python
fig, ax = plt.subplots()

# 5.5ポイントの幅の線で描画
ax.plot([1, 3], [3, 1], linewidth=5.5, label='5.5')
# 10ポイントの幅の線で描画
ax.plot([1, 3], [1, 3], linewidth=10, label='10')
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

# 破線で描画
ax.plot([1, 3], [3, 1], linestyle='--', label='dashed')
# 一点鎖線で描画
ax.plot([1, 3], [1, 3], linestyle='-.', label='dashdot')
# 点線で描画
ax.plot([1, 3], [2, 2], linestyle=':', label='dotted')
ax.legend()

plt.show()
```

##フォント
タイトル、凡例、軸ラベルなどのテキストに対してスタイル指定

```python
fig, ax = plt.subplots()

ax.set_xlabel('xlabel', family='fantasy', size=20, weight='bold')
ax.set_ylabel('ylabel', family='cursive', size=40, weight='light')
ax.set_title('graph title', family='monospace', size=25, weight='heavy')

plt.show()
```

```python
# フォントのスタイルを辞書で定義
fontdict = {
    'family' : 'fantasy',
    'size' : 20,
    'weight' : 'normal',
}

fig, ax = plt.subplots()

# 辞書形式でフォントのスタイルを指定
ax.set_xlabel('xlabel', fontdict=fontdict)
ax.set_ylabel('ylabel', fontdict=fontdict)
# 個別指定したsizeで上書き可能
ax.set_title('graph title', fontdict=fontdict, size=40)

plt.show()
```

##Pandasのオブジェクトからグラフ描画
PandasのDataframe、Seriesからグラフを描画できる
細かい表現の調整や複数のグラフの組み合わせることはできない

```python
import pandas as pd
import matplotlib.style
import matplotlib.pyplot as plt

matplotlib.style.use('ggplot') # スタイルを指定

# DataFram作成
df = pd.DataFrame({'A' : [1, 2, 3], 'B' : [3, 1, 2]})
df.plot() # 折れ線グラフを描画
plt.show()
```

```python
import numpy as np

# ランダムに３行２列のデータを生成してDataFrameを作成
np.random.seed(123)
df2 = pd.DataFrame(np.random.rand(3, 2),
                  columns=['y1', 'y2'])

df2.plot.bar() # 棒グラフを描画
plt.show()
```

```python
df2.plot.bar(stacked=True)
plt.show()
```
