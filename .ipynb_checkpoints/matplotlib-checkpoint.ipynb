##matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.style

# ggplotスタイル指定
matplotlib.style.use('ggplot')
```

**MATLAB-style**

```python
# データ用意
x = [1, 2, 3]
y = [2, 4, 9]

plt.plot(x, y) #折れ線グラフを描画
plt.title('MATLAB-style') #グラフのタイトル指定

plt.show() #グラフ表示
```

**OOP-style**

```python
# データ用意
x = [1, 2, 3]
y = [2, 4, 9]

# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

ax.plot(x, y) #折れ線グラフを描画
ax.set_title('OOP-style')

plt.show()
```

**描画オブジェクトとサブプロット**

```python
fig, axes = plt.subplots(2) # 2つのサブプロットを配置
plt.show()
```

```python
fig, axes = plt.subplots(2, 2) # 2行2列のサブプロットを配置
plt.show()
```

```python
fig, axes = plt.subplots(ncols=2) #1行2列のサブプロットを配置
plt.show()
```

**グラフのスタイル**

```python
print(matplotlib.style.available) # スタイルの一覧を表示
```

```python
# グラフのスタイルにclassicを指定
matplotlib.style.use('classic')

fig, ax = plt.subplots()
ax.plot([1, 2])

plt.show()
```

**タイトル**

```python
fig, axes = plt.subplots(ncols=2)

#　サブプロットにタイトル指定
axes[0].set_title('subplot title 0')
axes[1].set_title('subplot title 1')
#描画オブジェクトにタイトルを指定
fig.suptitle('figure title')

plt.show()
```

**軸ラベル**

```python
fig, ax = plt.subplots()

ax.set_xlabel('x label') #x軸にラベルを指定
ax.set_ylabel('y label') #y軸にラベルを指定

plt.show()
```

**判例**

```python
fig, ax = plt.subplots()

#凡例用のラベル設定
ax.plot([1, 2, 3],[2, 4, 9], label='legend label')
ax.legend(loc='best')

plt.show()
```

```python
fig, ax = plt.subplots()

#凡例用のラベル設定
ax.plot([1, 2, 3],[2, 4, 9], label='legend label')
ax.legend(loc='lower right') #凡例を右下に表示、10種類の中から選択可能。サブプロットの外側に表示したい場合は、bbox_to_anchor

plt.show()
```

**ファイル出力**

```python
# png、pdf、ps、eps、svgから選択可能。拡張子で判定
fig, ax = plt.subplots()
ax.set_title('subplot title')
fig.savefig('sample-figure.png') #png形式で保存
fig.savefig('sample-figure.svg') #svg形式で保存
```
